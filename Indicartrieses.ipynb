```python
from numpy import *
import matplotlib.pyplot as plt 

#make grid
dp = pi/400
theta, phi = mgrid[0:pi+dp:dp,0:2*pi+dp:dp]

#define functions
def f_sigma(theta, phi, beta):
    return (beta*sin(theta)-cos(phi))**2/(1-beta*sin(theta)*cos(phi))**5
f_sigma = vectorize(f_sigma)

def f_pi(theta, phi, beta):
    return cos(theta)**2 * sin(phi)**2/(1-beta*sin(theta)*cos(phi))**5
f_pi = vectorize(f_pi)
```

```python
#sigma plot
%matplotlib notebook
fig, axs = plt.subplots(nrows=1, ncols=3, subplot_kw=dict(projection='3d'), figsize=(10,3))

r = f_sigma(theta,phi,0.3)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[0].plot_surface(x,y,z)

axs[0].set_title(r'$f_\sigma,\beta=0.3$', fontsize = 14, fontweight = 'bold')
axs[0].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[0].set_ylabel('Y', fontsize = 12, fontweight = 'bold')



r = f_sigma(theta,phi,0.5)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[1].plot_surface(x,y,z)

axs[1].set_title(r'$f_\sigma,\beta=0.5$', fontsize = 14, fontweight = 'bold')
axs[1].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[1].set_ylabel('Y', fontsize = 12, fontweight = 'bold')


r = f_sigma(theta,phi,0.8)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[2].plot_surface(x,y,z)

axs[2].set_title(r'$f_\sigma,\beta=0.8$', fontsize = 14, fontweight = 'bold')
axs[2].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[2].set_ylabel('Y', fontsize = 12, fontweight = 'bold')
plt.savefig('s.png')
```

```python
#pi plot

fig, axs = plt.subplots(nrows=1, ncols=3, subplot_kw=dict(projection='3d'), figsize=(10,3))

r = f_pi(theta,phi,0.3)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[0].plot_surface(x,y,z)

axs[0].set_title(r'$f_\pi,\beta=0.3$', fontsize = 14, fontweight = 'bold')
axs[0].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[0].set_ylabel('Y', fontsize = 12, fontweight = 'bold')


r = f_pi(theta,phi,0.5)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[1].plot_surface(x,y,z)

axs[1].set_title(r'$f_\pi,\beta=0.5$', fontsize = 14, fontweight = 'bold')
axs[1].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[1].set_ylabel('Y', fontsize = 12, fontweight = 'bold')


r = f_pi(theta,phi,0.8)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[2].plot_surface(x,y,z)

axs[2].set_title(r'$f_\pi,\beta=0.8$', fontsize = 14, fontweight = 'bold')
axs[2].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[2].set_ylabel('Y', fontsize = 12, fontweight = 'bold')


plt.savefig('pi.png')
```

```python
#sum plot

fig, axs = plt.subplots(nrows=1, ncols=3, subplot_kw=dict(projection='3d'), figsize=(10,3))

r = f_sigma(theta,phi,0.3) + f_pi(theta,phi,0.3)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[0].plot_surface(x,y,z)

axs[0].set_title(r'$f_\sigma + f_\pi,\beta=0.3$', fontsize = 14, fontweight = 'bold')
axs[0].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[0].set_ylabel('Y', fontsize = 12, fontweight = 'bold')

r = f_sigma(theta,phi,0.5)+f_pi(theta,phi,0.5)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[1].plot_surface(x,y,z)

axs[1].set_title(r'$f_\sigma+f_\pi,\beta=0.5$', fontsize = 14, fontweight = 'bold')
axs[1].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[1].set_ylabel('Y', fontsize = 12, fontweight = 'bold')


r = f_sigma(theta,phi,0.8)+f_pi(theta,phi,0.8)
x = r * sin(theta)*cos(phi)
y = r * sin(theta)*sin(phi)
z = r * cos(theta)


axs[2].plot_surface(x,y,z)

axs[2].set_title(r'$f_\sigma + f_\pi,\beta=0.8$', fontsize = 14, fontweight = 'bold')
axs[2].set_xlabel('X', fontsize = 12, fontweight = 'bold')
axs[2].set_ylabel('Y', fontsize = 12, fontweight = 'bold')

plt.savefig('sum.png')
```
